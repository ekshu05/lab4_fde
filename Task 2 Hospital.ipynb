```python
from pyspark.sql.functions import sum as spark_sum

# Load from Silver layer
df_cleaned = spark.table("silver_hospital_orders")

# Revenue by city (instead of department)
df_city_sales = (
    df_cleaned.groupBy("city")
             .agg(spark_sum("total_value").alias("total_revenue"))
)

df_city_sales.createOrReplaceTempView("tmp_hospital_city_sales")

spark.sql("""
CREATE OR REPLACE TABLE gold_hospital_city_sales AS
SELECT * FROM tmp_hospital_city_sales
""")

# Daily revenue
df_daily_sales = (
    df_cleaned.groupBy("order_date")
             .agg(spark_sum("total_value").alias("daily_revenue"))
)

df_daily_sales.createOrReplaceTempView("tmp_hospital_daily_sales")

spark.sql("""
CREATE OR REPLACE TABLE gold_hospital_daily_sales AS
SELECT * FROM tmp_hospital_daily_sales
""")
```
